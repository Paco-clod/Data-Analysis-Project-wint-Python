```python
# Importing pandas library
import pandas as pd

# Given URL to the dataset
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMSkillsNetwork-AI0272EN-SkillsNetwork/labs/dataset/2016.csv"

# Function to load the data and check column data types
def check_data_types(file_url):
    # Reading the CSV file into a DataFrame
    df = pd.read_csv(file_url, header=0)

    # Checking and displaying column data types
    print("\nData Types of Each Column:")
    print(df.dtypes)

    # Define expected data types
    expected_types = {
        'Country': 'object',
        'Region': 'object',
        'Happiness Rank': 'int64',
        'Happiness Score': 'float64',
        'Lower Confidence Interval': 'float64',
        'Upper Confidence Interval': 'float64',
        'Economy (GDP per Capita)': 'float64',
        'Family': 'float64',
        'Health (Life Expectancy)': 'float64',
        'Freedom': 'float64',
        'Trust (Government Corruption)': 'float64',
        'Generosity': 'float64',
        'Dystopia Residual': 'float64'
    }

    # Verifying each column data type
    for col, col_type in expected_types.items():
        if df[col].dtype != col_type:
            print(f"\nColumn '{col}' has unexpected type. Expected: {col_type}, Found: {df[col].dtype}")
        else:
            print(f"Column '{col}' has the correct type: {col_type}")

# Calling the function with the provided URL
check_data_types(URL)
```

```python
# Function to clean the DataFrame
def clean_dataframe(file_url):
    # Reading the CSV file into a DataFrame
    df = pd.read_csv(file_url, header=0)

    # Task 1: Remove leading and trailing whitespaces from a column, say 'Country'
    for col in ['Country', 'Region']:  # Assuming 'Country' and 'Region' might have whitespaces
        df[col] = df[col].str.strip()
        print(f"Removed whitespaces from column {col}.")

    # Task 2: Replace empty strings with NaN
    for col in ['Country', 'Region']:  # Assuming 'Country' and 'Region' might have empty strings
        df[col] = df[col].replace('', pd.NA, regex=True)
        print(f"Replaced empty strings with NaN in column {col}.")

    # Task 3: Convert data types to appropriate types
    # Assuming directives from our initial data type check
    # Converting columns to numeric, handling errors by coercing to NaN
    numeric_cols = ['Happiness Rank', 'Happiness Score', 'Lower Confidence Interval',
                    'Upper Confidence Interval', 'Economy (GDP per Capita)', 'Family',
                    'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)',
                    'Generosity', 'Dystopia Residual']
    for col in numeric_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce') # This line is changed to handle errors

    print("Data types updated in the DataFrame.")

# Calling the function with the provided URL
clean_dataframe(URL)
```

```python
# Function to handle missing values
def handle_missing_values(file_url):
    # Reading the CSV file into a DataFrame
    df = pd.read_csv(file_url, header=0)

    # Task 1: Identify columns with missing values
    missing_values = df.isnull().sum()
    print("\nColumns with Missing Values:")
    print(missing_values[missing_values > 0])

    # Task 2: Replace missing values with column mean
    columns_to_impute = missing_values[missing_values > 0].index
    for col in columns_to_impute:
        # Convert the column to numeric before calculating the mean, handling errors by coercing to NaN
        df[col] = pd.to_numeric(df[col], errors='coerce')
        mean_value = df[col].mean()
        df[col].fillna(mean_value, inplace=True)
        print(f"Missing values in column '{col}' replaced with mean: {mean_value}")

    # Print the DataFrame to check changes
    print("\nDataFrame after handling missing values:")
    print(df.head())

# Calling the function with the provided URL
handle_missing_values(URL)
```

```python
import pandas as pd
import plotly.express as px

# Function to create bar chart of top 10 countries
def create_top_10_bar_chart():
    # Reading the CSV file into a DataFrame
    df = pd.read_csv(URL, header=0)

    # Task 1: Select top 10 countries by Happiness Score
    top_10_countries = df.sort_values(by='Happiness Score', ascending=False).head(10)

    # Task 2: Extract GDP per Capita and Healthy Life Expectancy for these countries
    gdp_per_capita = top_10_countries['Economy (GDP per Capita)']
    healthy_life_expectancy = top_10_countries['Health (Life Expectancy)']

    # Creating a DataFrame for the bar chart
    data_for_chart = pd.DataFrame({
        'Country': top_10_countries['Country'],
        'GDP per Capita': gdp_per_capita,
        'Healthy Life Expectancy': healthy_life_expectancy
    })

    # Setting up the Plotly figure
    fig1 = px.bar(data_for_chart, x='Country', y=['GDP per Capita', 'Healthy Life Expectancy'],
                 title='GDP per Capita and Healthy Life Expectancy of Top 10 Happiest Countries',
                 labels={'GDP per Capita': 'GDP per Capita (USD)', 'Healthy Life Expectancy': 'Healthy Life Expectancy (years)'},
                 barmode='group')

    # Show the figure
    fig1.show()

# Call the function to create the chart
create_top_10_bar_chart()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Function to create correlation heatmap
def create_correlation_heatmap():
    # Reading the CSV file into a DataFrame
    df = pd.read_csv(URL, header=0)

    # Replace empty strings and spaces with NaN (for numeric columns)
    for col in ['Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)',
                'Freedom', 'Trust (Government Corruption)', 'Generosity', 'Happiness Score']:
        df[col] = df[col].replace(r'^\s*$', pd.NA, regex=True)

    # Convert columns to numeric
    numeric_cols = ['Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)',
                    'Freedom', 'Trust (Government Corruption)', 'Generosity', 'Happiness Score']
    for col in numeric_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce')

    # Subset of required columns
    sub_df = df[numeric_cols]

    # Compute correlation matrix
    corr_matrix = sub_df.corr()

    # Create heatmap with annotations
    fig2 = go.Figure(data=go.Heatmap(
        z=corr_matrix.values,
        x=corr_matrix.columns,
        y=corr_matrix.index,
        colorscale='Bluered',
        zmin=-1,
        zmax=1,
        colorbar=dict(title='Correlation Coefficient')
    ))

    # Add correlation value annotations in each cell
    for i in range(len(corr_matrix)):
        for j in range(len(corr_matrix.columns)):
            fig2.add_annotation(
                x=corr_matrix.columns[j],
                y=corr_matrix.index[i],
                text=str(round(corr_matrix.iloc[i, j], 2)),
                showarrow=False,
                font=dict(
                    color="white" if abs(corr_matrix.iloc[i, j]) > 0.5 else "black"
                )
            )

    # Update layout
    fig2.update_layout(
        title='Correlation Heatmap for Key Determinants of Happiness',
        xaxis_title='Attributes',
        yaxis_title='Attributes',
        width=800,
        height=600
    )

    # Show the heatmap
    fig2.show()

# Call the function to generate the heatmap
create_correlation_heatmap()
```

```python
# Function to create scatter plot
def create_gdp_happiness_scatter():
    # Reading the CSV file into a DataFrame
    df = pd.read_csv(URL, header=0)

    # Step 1: Create a subset for plotting, including only relevant columns
    plot_df = df[['Happiness Score', 'Economy (GDP per Capita)', 'Region']]

    # Step 2: Use Plotly Express to create the scatter plot
    fig3 = px.scatter(plot_df, x='Economy (GDP per Capita)', y='Happiness Score', color='Region',
                      title='Relation between GDP per Capita and Happiness Score by Region',
                      labels={'Economy (GDP per Capita': 'GDP per Capita (USD)',
                              'Happiness Score': '2016 Happiness Score'},
                      width=800,
                      height=600)

    # Show the scatter plot
    fig3.show()

# Call the function to generate the scatter plot
create_gdp_happiness_scatter()
```

```python
# Function to create pie chart
def create_happiness_by_region_pie_chart():
    # Reading the CSV file into a DataFrame
    df = pd.read_csv(URL, header=0)

    # Step 1: Aggregate 'Happiness Score' by 'Region'
    region_happiness = df.groupby('Region')['Happiness Score'].agg(['mean']).reset_index()

    # Step 2: Create a Plotly pie chart
    fig4 = px.pie(region_happiness, names='Region', values='mean', color='Region',
                   title='Distribution of Happiness Score by Region',
                   labels=dict(mean='Average Happiness Score'))

    # Customize chart appearance
    fig4.update_traces(textposition='inside', textinfo='percent+label')
    fig4.update_layout(width=600, height=500)

    # Show the pie chart
    fig4.show()

# Call the function to generate the pie chart
create_happiness_by_region_pie_chart()
```

```python
# Function to create GDP per Capita map with Health Life Expectancy tooltip
def create_gdp_life_expectancy_map():
    # Reading the CSV file into a DataFrame
    df = pd.read_csv(URL, header=0)

    # Load country shapes data using gapminder dataset
    shapes = px.data.gapminder().query("year == 2007").reset_index() # Using gapminder and selecting year 2007 for demonstration
    shapes['id'] = shapes['country'].apply(lambda x: x.lower().replace(" ", "_"))

    # Merge GDP and Life Expectancy data with country shapes
    merged_data = pd.merge(shapes, df[['Country', 'Economy (GDP per Capita)', 'Health (Life Expectancy)']],
                           left_on='country', right_on='Country', how='left') # Changed 'on' to 'left_on' and 'right_on'


    # Convert 'Economy (GDP per Capita)' to numeric before calculating min/max
    # This ensures that the range_color argument gets numeric values
    df['Economy (GDP per Capita)'] = pd.to_numeric(df['Economy (GDP per Capita)'], errors='coerce')

    # Create a multiplex plot (choropleth map)
    fig5 = px.choropleth(merged_data, locations="iso_alpha", # Changed location to iso_alpha
                        locationmode='ISO-3', # Changed location mode
                        color='Economy (GDP per Capita)',
                        hover_name='Health (Life Expectancy)',
                        title='GDP per Capita per Country with Health Life Expectancy Tooltip',
                        # animation_frame="Year",  # Commenting out animation frame as 'Year' column is not available
                        projection='natural earth',
                        color_continuous_scale=px.colors.sequential.Plasma,
                        range_color=[df['Economy (GDP per Capita)'].min(), df['Economy (GDP per Capita)'].max()]
                        )

    # Update layout for better visualization
    fig5.update_layout(
        title_text="Global GDP per Capita with Health Life Expectancy Tooltip",
        geo_scope='world',
        paper_bgcolor='rgba(0,0,0,0)',
        plot_bgcolor='rgba(0,0,0,0)'
    )

    # Show the map
    fig5.show()

# Call the function to generate the map
create_gdp_life_expectancy_map()
```
